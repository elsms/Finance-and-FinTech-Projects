# Importing the data

The dataset was found here: https://www.kaggle.com/datasets/pratyushpuri/crypto-market-sentiment-and-price-dataset-2025

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('crypto_dataset.csv')
data.head()
```

# Exploratory Data Anaysis (EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_subset=data.iloc[:,2:]

plt.figure(figsize=(12, 10))
sns.heatmap(num_subset.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
len(num_subset.columns)
```

```python
list_colors = ['orangered', 'yellow', 'lawngreen', 'palegreen', 'aquamarine',
               'dodgerblue', 'slategray', 'royalblue', 'slateblue', 'indigo',
               'magenta', 'crimson']

for idx, i in enumerate(num_subset.columns):
    plt.hist(num_subset[i], bins=10, edgecolor='black', color=list_colors[idx % len(list_colors)])
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.show()
```

```python
plt.bar(x=data['cryptocurrency'], y=data['social_sentiment_score'], height=12)
plt.title('Social Sentiment Score by cryptocurrency')
plt.xlabel('Cryptocurrency')
plt.ylabel('Social Sentiment Score')
plt.show()
```
